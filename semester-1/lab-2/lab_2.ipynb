### Задание 1.2
- **1 вариант**: пользуясь материалами лекции, реализовать алгоритм $Apriori$ и для своей подвыборки сгенерировать ассоциативные правила с уровнем уверенности не менее **50%** (уровень поддержки для часто встречающиеся подмножества объектов не менее **30%**).
- **2 вариант**: пользуясь материалами лекции, реализовать алгоритм $FP-Growth$ и для своей подвыборки найти часто встречающиеся подмножества объектов с уровнем поддержки не менее $2$.

```python
from itertools import chain, combinations
```

```python
def support(subset, data):
    return sum([1 for item in data if set(subset).issubset(item)])
```

```python
def powerset(s):
    s = list(s)
    return list(set(x) for x in chain.from_iterable(
        combinations(s, r) for r in range(len(s) + 1)))
```

```python
class Rule(object):
    def __init__(self, antecedent, consequent):
        self._antecedent = antecedent
        self._consequent = consequent
        
    def __str__(self):
        return f'{self._antecedent} => {self._consequent}'
```

```python
def generate_rules(frequent_subsets, data, min_confidence):
    result = []
    
    for frequent_subset in frequent_subsets:
        for subset in powerset(frequent_subset)[1:-1]:
            antecedent = subset
            consequent = frequent_subset - subset
            
            confidence = support(frequent_subset, data) / support(subset, data)
            if confidence >= min_confidence:
                result.append(Rule(antecedent, consequent))
                
    return result
```

----

```python
from collections import defaultdict
from itertools import chain, combinations

class Node:
    def __init__(self, name, frequency, parent):
        self.name = name
        self.count = frequency
        self.parent = parent
        self.children = {}
        self.next = None

def construct_fp_tree(subsets, min_support, frequency=None):
    frequency = [1 for _ in subsets] if frequency is None else frequency
    items = defaultdict(int)
    
    for idx, subset in enumerate(subsets):
        for item in subset:
            items[item] += frequency[idx]

    items = {item: support for item, support in items.items() if support >= min_support}
    if(len(items) == 0):
        return None

    for item in items:
        items[item] = [items[item], None]

    tree = Node('null', 1, None)
    for idx, subset in enumerate(subsets):
        subset = [item for item in subset if item in items]
        subset.sort(key=lambda item: items[item][0], reverse=True)
        node = tree
        for item in subset:
            node = update_tree(item, node, items, frequency[idx])

    return items

def update_items(item, target, items):
    if(items[item][1] is None):
        items[item][1] = target
    else:
        node = items[item][1]
        while node.next is not None:
            node = node.next
        node.next = target

def update_tree(item, node, items, frequency):
    if item in node.children:
        node.children[item].count += frequency
    else:
        new_node = Node(item, frequency, node)
        node.children[item] = new_node
        update_items(item, new_node, items)

    return node.children[item]

def collect_prefix_path(node, prefix_path):
    if node.parent is not None:
        prefix_path.append(node.name)
        collect_prefix_path(node.parent, prefix_path)

def prefix_path(item, items):
    node = items[item][1] 
    path_base, frequency = [], []
    
    while node != None:
        prefix_path = []
        collect_prefix_path(node, prefix_path)  
        
        if len(prefix_path) > 1:
            path_base.append(prefix_path[1:])
            frequency.append(node.count)

        node = node.next
    return path_base, frequency

def mine_tree(items, min_support, frequent_items, prefix=None):
    prefix = set() if prefix is None else prefix
    
    sorted_items = [item[0] for item in sorted(items.items(), 
                                               key=lambda p: p[1][0])]
    
    for item in sorted_items:  
        new_prefix = prefix.copy()
        new_prefix.add(item)
        frequent_items.append(new_prefix)
        
        path_base, frequency = prefix_path(item, items) 
        newitems = construct_fp_tree(path_base, min_support, frequency) 
        
        if newitems is not None:
            mine_tree(newitems, min_support, frequent_items, new_prefix)
```

```python
def fp_growth(dataset, min_support, min_confidence):
    min_support *= len(dataset)
    items = construct_fp_tree(dataset, min_support)
    
    frequent_items = []
    mine_tree(items, min_support, frequent_items)
    
    return generate_rules(frequent_items, dataset, min_confidence)
```

```python
for idx, rule in enumerate(fp_growth(sample, 0.3, 0.5)):
    print(f'Rule {idx + 1}: \t{rule}')
```

## Задание 2 (1 балл)
Реализовать алгоритм для другого варианта и продемонстрировать его работу на Вашей подвыборке.

```python
def join(subsets):
    items = {item for item in chain(*subsets)}
    
    result = []
    for subset in subsets:
        for new_item in items:
            new_subset = subset.copy()
            new_subset.add(new_item)
            if len(new_subset) > len(subset) and new_subset not in result:
                result.append(new_subset)
                
    return result
```

```python
def apriori(dataset, min_support, min_confidence):
    I = {name for name in chain(*dataset)}
    
    C = [set([x]) for x in I]
    L = [x for x in C if support(x, dataset) / len(dataset) >= min_support]
    
    result = []
    
    while L != list():
        L = [x for x in join(L) if support(x, dataset) / len(dataset) >= min_support]
        result += L
        
    return generate_rules(result, dataset, min_confidence)
```

```python
for idx, rule in enumerate(apriori(sample, 0.3, 0.5)):
    print(f'Rule {idx + 1}: \t{rule}')
```
